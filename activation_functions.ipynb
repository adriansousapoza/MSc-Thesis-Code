```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Define a range of values for the x-axis
x_values = torch.linspace(-10, 10, 500)

# Compute the activation function values
activations = {
    "Sigmoid": torch.sigmoid(x_values),
    "ReLU": torch.relu(x_values),
    "Tanh": torch.tanh(x_values),
    "Softplus": torch.nn.functional.softplus(x_values),
    "ELU": torch.nn.functional.elu(x_values, alpha=1.0),
    "GELU": torch.nn.functional.gelu(x_values),
    "Swish": x_values * torch.sigmoid(x_values),
    "PReLU": torch.nn.PReLU()(x_values)
}

# Define line styles and colors
line_styles = ['-', '--', '-.', ':', '-', '--', '-.', ':']
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange']

# Detach tensors to disable gradient tracking for plotting
x_values_np = x_values.detach().numpy()

# Compute the activation function values again without gradients
activations_np = {name: values.detach().numpy() for name, values in activations.items()}

# Plot each activation function with distinct line styles and colors
plt.figure(figsize=(10, 8))
for i, (name, values) in enumerate(activations_np.items()):
    plt.plot(x_values_np, values, label=name, linestyle=line_styles[i], color=colors[i])

# Add labels, title, and legend
plt.title("Common Activation Functions in Neural Networks")
plt.xlabel("Input")
plt.ylabel("Activation Output")
plt.legend()
plt.grid(True)
plt.ylim(-2, 2)
plt.xlim(-5, 5)
plt.savefig("figures/activation_functions.pdf", format="pdf", dpi=300, bbox_inches="tight")
plt.show()
```
